```python
import numpy as np
import pandas as pd
```

```python
# On defini d'abord l'url du site qu'on veut extraire les tableau

url = 'https://en.wikipedia.org/wiki/World_population'
```

```python
tables = pd.read_html(url)
```

```python
# nombre de tableaux html trouves a la page de cette url
len(tables)
```

```python
# afficher le  tableau parmis les 30
tables[7]
```

```python
# sauvegarde du df en tant qu'un fichier html
tables = tables[7]
tables.to_html('fichier.html', index=False)
```
