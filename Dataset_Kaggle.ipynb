Travail fait sur le dataset Premier League Match Data avec **Seaborn**

Présentation du dataset :

Le dataset contient les statistiques des matchs de chaque rencontre et de chaque équipe de la saison 2021-2022 de la Premier League anglaise. Il inclut des informations telles que les équipes, les arbitres, et les statistiques individuelles des matchs à domicile et à l'extérieur, telles que les fautes, les tirs et les cartons. De plus, il comprend également un ensemble de données sur les classements hebdomadaires de la saison.

Messages clés du dataset :

La saison 2021-2022 de la Premier League était la 30e saison de la Premier League et la 123e saison du football anglais de premier plan dans son ensemble.
Les dates de début et de fin de la saison ont été annoncées le 25 mars 2021, avec les calendriers publiés le 16 juin 2021.
Manchester City a réussi à défendre son titre en remportant un sixième titre de Premier League, marquant ainsi leur huitième titre de championnat anglais au total.



Histoire à raconter :

La saison 2021-2022 de la Premier League a été marquée par la domination de Manchester City, qui a réussi à défendre son titre de champion. Malgré la concurrence féroce des autres équipes, Manchester City a maintenu sa position dominante et a remporté un sixième titre de Premier League, consolidant ainsi sa place parmi les meilleures équipes de l'histoire de la Premier League.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/soccer21-22.csv')
df
```

```python
df.head()
```

Visualisation en utilisant la  fonction de personnalisation de Matplotlib
---



```python



sns.lineplot(x="HomeTeam", y="FTHG", data=df)
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
plt.show()
```

Visualisation en utilisant la  methode scatterplot()


```python
sns.scatterplot(x='HomeTeam', y='FTHG', data=df,)
plt.show()
```

Visualisation aen utilisant la methode lineplot()

```python
sns.lineplot(x='HomeTeam', y='FTHG', data=df)
plt.show()
```

Visualisation  avec Bokeh en  utilisant la methode line()

```python
# using only data attribute
sns.lineplot(data=df.drop(['HomeTeam'], axis=1))
plt.show()
```

Visualisation en utilisant la methode barplot()

```python
sns.barplot(x='HomeTeam',y='FTHG', data=df, hue='AwayTeam')
plt.show()
```

Construction de l'histogramme  dans seaborn en utililant  la methode barplot()

```python
sns.histplot(x='HomeTeam', data=df, kde=True,
hue='AwayTeam')
plt.show()
```
