```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/MasterClass_DataScience_Machine_Learning_Notebooks_et_Datasets/05-Seaborn/Datasets/dm_office_sales.csv')
df
```

```python
''' Nous avons 3 diagrammes de distribution seaborn que sont:
                  # rugplot()
                      il est souvent considere comme diagramme de distribution a une dimension
                      ici l'axe y n'est pas interpretable donc n'a pas d'interet


              pour les histogrammes , on a:
                # displot() :  qui est une version generalisee de histplot()

                # histplot()

'''
```

```python

fig = plt.figure(figsize=(10,9), dpi=200)

sns.rugplot(x='salary',
            data=df, # notre dataset
            height=0.5, # hauteur
            );

plt.savefig( 'diagram de dispersion avec rugplot.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(10,9), dpi=200)

sns.set(style='darkgrid' # le style pour l'aujout de la couleur d'arriere plan de la grille. Les options, on peut avoir
                # whitegrid
                # white : qui est le style par defaut
                # dart
                # ticks
        )
sns.displot(data=df,
            x='salary',
            bins=60 ,# pour le nombre de bars
            kde=True # estimation de la densite du noyau: karnel density estimation
            rug=True # Pour le diagramme rugplot

            );
```

```python
fig = plt.figure(figsize=(10,10), dpi=200)
sns.histplot(data=df,
             x='salary',
             kde=True,
             )
```

```python
# Courbe d'estimation de la densite du noyau: Kernel density estimation kde
#fig = plt.figure(figsize=(8,8), dpi=200)

sns.kdeplot(data=df,
            x='salary',

            );
```

```python
# greation d'1 dataframe a partir des nombres etiers aleatoires
np.random.seed(42)
sample_ages = np.random.randint(0,100,100)
df = pd.DataFrame(sample_ages, columns=['age'])
```

```python
df
```

```python
# Diagram rugplot
sns.rugplot(df, x='age');
```

```python
# histogram incluant les diagram rugplot et kdeplot

sns.displot(df, x='age', rug=True, bins=30, kde=True);
```

```python
# Diagram kde
sns.kdeplot(data=df,
            x='age',
            clip=[0,100],
            bw_adjust=0.02, # Pour band weight : bande passante
            shade=True # remplir en-dessous de la courbe

            )
```
