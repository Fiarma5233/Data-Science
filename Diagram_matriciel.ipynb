```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/MasterClass_DataScience_Machine_Learning_Notebooks_et_Datasets/05-Seaborn/Datasets/country_table.csv')
df
```

```python
# on  prends comme index, les continents
df = df.set_index('Countries')
df
```

```python
# Construction du diagram heatmap : diagramm thermique
fig = plt.figure(figsize=(10,8), dpi=200 )
sns.heatmap(
    df.drop('Life expectancy', axis=1),  # on suprime la colonne  car elle a une unite diferente des autres colonnes
    lw=0.7, # Pour un espacement
    annot=True, # Pour ajouter une anotation
    cmap='viridis' # pour la  couleur. ici on rend cmap au lieu de palette
)
```

```python
# Construction du diagram heatmap : diagramm thermique
fig = plt.figure(figsize=(10,8), dpi=200 )
sns.clustermap(
    df.drop('Life expectancy', axis=1),  # on suprime la colonne  car elle a une unite diferente des autres colonnes
    lw=0.7, # Pour un espacement
    annot=True, # Pour ajouter une anotation
    cmap='viridis' # pour la  couleur. ici on rend cmap au lieu de palette
)
```
