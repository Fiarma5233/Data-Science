```python


import numpy as np
import pandas as pd

# %% [markdown]
## Concat√©nation


data_one = {'A': ['A0', 'A1', 'A2', 'A3'],'B': ['B0', 'B1', 'B2', 'B3']}

# %% [code]
data_two = {'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']}

# %% [code]
one = pd.DataFrame(data_one)

# %% [code]
two = pd.DataFrame(data_two)

```

```python
one
```

```python
two
```

```python
# concatenation de 2 dataframes sur la colonne
''' La metode concat() prend en parametre une listes des df a concatener puis l'axis (le long des colonnes) axis=1

'''

pd.concat([one, two], axis=1)
```

```python
# La concatenation sur les lignes n'a vraiement pas de sens car ca donne le plus souvent ceci
pd.concat([one, two], axis=0)

# Pour le faire, il faut d'abord renommer les colonnes de la df two
```

```python
# pour renommer les colones de la df two en gardant toujours les memes valeurs, on a:
two.columns = one.columns
two
```

```python
# ensuite on concatene
pd.concat([one, two], axis=0)
```

```python
# on remarque les index ont ete dupliques, pour resoudre ceci, on donne d'abord un nom a notre df concatene
my_df_concat = pd.concat([one, two], axis=0)
my_df_concat
```

```python
# ensuite on modifie les index
my_df_concat.index = range(len(my_df_concat))
my_df_concat
```

    Fusion avec la methode merge()**texte en gras** **texte en gras**

Tables des donnees

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
help(pd.merge)
```

```python
pd.merge(registrations, logins, how='inner', on='name')
'''
  registrations = table a gauche
  logins = table a droite
      Remarque : L'ordre dans lequel on passe les tables dans la fonction merge n'a pas d'importance,
  how = comment on veut fusioner ces tables
  inner = jointure interne ( prend les elements presents a la fois dans les 2 tables)
  on = sur
  name = veut ire que c'est sur le nom qu'on souhaite faire

'''
```

```python
# Fusion a gauche

'''
  left = registrations = table a gauche
  right = logins = table a droite
      Remarque : L'ordre dans lequel on passe les tables dans la fonction merge a une grande importance,
  how = comment on veut fusioner ces tables
  left = jointure a gauche ( tous les elements de la df gauche seront presents dans les resultats)
  on = sur
  name = veut ire que c'est sur le nom qu'on souhaite faire

'''
# ie ans la fusion gauche, on cherche seulement les elements qui sont presnt dans la df gauche

pd.merge(left=registrations, right=logins, how='left', on='name')


```

```python
# Fusion a droite

'''
  left = registrations = table a gauche
  right = logins = table a droite
      Remarque :
  how = comment on veut fusioner ces tables
  right = jointure a gauche ( tous les elements de la df droite seront presents dans les resultats)
  on = sur
  name = veut ire que c'est sur le nom qu'on souhaite faire

'''
# ie ans la fusion gauche, on cherche seulement les elements qui sont presnt dans la df gauche

pd.merge(left=registrations, right=logins, how='right', on='name')
```

```python
# Jointure externe

'''
  registrations = table a gauche
  logins = table a droite
      Remarque : L'ordre dans lequel on passe les tables dans la fonction merge n'a pas d'importance,
  how = comment on veut fusioner ces tables
  outer = jointure externe ( prend les elements presents dans les 2 tables)
  on = sur
  name = veut ire que c'est sur le nom qu'on souhaite faire

'''
```

```python
pd.merge(registrations, logins, how='outer', on='name')
```
