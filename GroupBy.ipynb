```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/MasterClass_DataScience_Machine_Learning_Notebooks_et_Datasets/03-Pandas/Datasets/mpg.csv')
df
```

```python
df.head()
```

```python
# on utilise le groupby() sur des colonnes categorielles(qui peuvent etre des chaines de caracteres ou numerique)
```

```python
# groupby avec une seule colonne
df.groupby('model_year').mean()
```

```python
# groupby avec plusieurs colonnes ou index multiniveaus
df.groupby(['model_year','cylinders']).mean().index

```

```python
# afficher la moyenne pour chq anne et ayant un cylindre 6 ou 8
df[df['cylinders'].isin([6,8])].groupby(['model_year', 'cylinders']).mean()
```

```python
# methode pour specifier un groupe de fonction d'aggregation: agg()
df.agg(['std', 'sum'])['mpg']
```

```python
df.head()
```

```python
# on peut le faire en utilisant un dictionnaire qui aura comme cle les noms de colonnes et valeurs les fonctuons d'agregation qu'on veut
# groupby avec plusieurs colonnes ou index multiniveau
df.agg({'mpg':['mean', 'sum'], 'weight':['var', 'std'], 'acceleration':['min', 'max'] })
```
