```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
a
```

```python
b = a ** 4
b
```

```python
x = np.arange(0,10)
x
```

```python
y = 2 * x
y
```

```python
fig = plt.figure(dpi=200,figsize=(5, 4)) # creation du canvas de la figure
# dpi = pouces par pixels pour la resolution
# figsize pour la taille de la figure

#================== Grande figure Axe 1 ===================================>
# ajout d'axes
  #  ces valeurs [0,0,1,1] correspondent respectivement :
            # Point inferieur gauche de l'axe des abscisse
            # Point inferieur gauche de l'axe des ordonnees
                      #ie que ces 2 premiers points sont les coordonnees du points inferieur gauche
            # La largeur
            # La hauteur


axes1 = fig.add_axes([0,0,1,1])
# tracage d'une droite
axes1.plot(a,b); # le ";" a la fin ici permet de supprimer l'affichage au dessus de la figure indiquant la taille memoire et autre

# ajouter des limitations sur notre grande fig d'axe1
axes1.set_xlim(0,9)
axes1.set_ylim(0,40)

# ajouter des label a  notre grande fig d'axe1
axes1.set_xlabel('A')
axes1.set_ylabel('B')


# ajouter u titre a notre grande fig d'axe1
axes1.set_title('Puissance de 4')


#================== Petite figure  Axe 2 qui sera inseree dans la grande ===================================>
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(x,y)

 #ajouter des limitations sur notre grande fig d'axe1
axes2.set_xlim(0,1)
axes2.set_ylim(0,1)

# ajouter des label a  notre grande fig d'axe1
axes2.set_xlabel('X')
axes2.set_ylabel('Y')


# ajouter u titre a notre grande fig d'axe1
axes2.set_title('Zoom en avant')

plt.savefig('Figure incorporee', bbox_inches='tight')
# pour garder  les nomminations des axes lorqu'on enregistre
```
