```python
import numpy as np
import pandas as pd
import datetime as datetime
```

```python
my_series= pd.Series(['Feb 3, 2022', '2001-12-31', None])
```

```python
my_series
```

```python
pd.to_datetime(my_series, format='mixed') # Pour convertir en date
```

```python
#Il ya deux formats de date:
      #Le format europeen : annee-mois-jour
obvious_euro_date = "31-12-2001" # version francaise comme on l'ecris chez nous
pd.to_datetime(obvious_euro_date) # conversion en date qui par defaut est au format americain


#Le format americain : annee-jour-mois
euro_date = "10-09-2009"
pd.to_datetime(euro_date)


# Pour convertir une date dans le format europeen , on specifie en parametre : dayfirst=True
pd.to_datetime(euro_date, dayfirst=True)

```

```python
# specifier le format de date d'entree
date_bizarre = '23--Feb--2024'
pd.to_datetime(date_bizarre, format='%d--%b--%Y')
```
