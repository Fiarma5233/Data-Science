```python
# on definis la fonction qui prendra comme parametres les colonnes qu'on va utiliser
def quality(total_bill, tip):
  if tip / total_bill > 0.25:
    return "Genereux"

  else:
    return "Another"
```

```python
quality(16.99, 1.01)
```

```python
# Methode generale d'utilisation de lambda pour les colonnes multiples
  # on selection les colonnes de la dataframe
  # on applique la methode apply()
  # on utilise la fonction lambda dans la methode apply()
  # la methde apply() prend 2 parametres que sont : la fonction lambda et axis=1

# voici l'exemple
df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
# pour utiliser ca comme une nouvelle colonne :
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
df
```

```python
# on peut aussi utiliser np.vectorize a la place de lambda
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
# pour remplacer des valeurs dans la serie , on peut utiliser 2 methodes a savoir:
  # replace() ### Lorsqu'il s'agit de 1 ou 2 valeurs a remplacer
  # map() ### Lorsque c'est plusieurs valeurs, on creee d'abord un dictionnaire puis on map
```

```python
# methode replace() avec 1 valeur
df['sex'].replace(to_replace="Male", value="M") # Pour remplacer une seule valeur de la serie
```

```python
# methode replace() avec 2 valeurs
df['sex'].replace(to_replace=["Male", "Female"], value=["M", "F"]) # Pour remplacer plusieurs valeurs de la serie
```

```python
# Avec la methode map()

my_map = {'Female':"F", 'Male':"M"} # creation du dictionnaire

df['sex'].map(my_map) # mapping
```
