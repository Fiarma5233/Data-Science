Représentation graphique :

```python
import numpy as np
import pandas as pd
import plotly.express as px

```

```python
df = pd.read_csv('/content/drive/MyDrive/MasterClass_DataScience_Machine_Learning_Notebooks_et_Datasets/02-NumPy/netflix1.csv')
```

```python
df.head()
```

Visualisation en utilisant la methode line()

```python

# plotting the scatter chart
fig = px.line(df, y='type', color='country')
# showing the plot
fig.show()
```

```python
df.head()
```

Visualisation en utilisant la methode scatter() de plotly.express.

```python
# plotting the scatter chart
fig = px.scatter(df, x="country", y="type", color='date_added')
# showing the plot
fig.show()
```

Plotly
Bar Chart
Le graphique à barres dans Plotly peut être créé en utilisant la méthode bar()
de la classe plotly.express.

```python
# plotting the scatter chart
fig = px.line(df, y='type', color='date_added')
# showing the plot
fig.show()
```

Visualisation du graphique à barres dans Plotly  en utilisant la méthode bar()
de la classe plotly.express.

```python
# plotting the scatter chart
fig = px.bar(df, x='country', y='type', color='date_added')
# showing the plot
fig.show()
```

Visualisation des histogrammes à l'aide de la fonction
histogram() de la classe plotly.express.

```python
# plotting the scatter chart
fig = px.histogram(df, x='country', color='date_added')
# showing the plot
fig.show()
```
