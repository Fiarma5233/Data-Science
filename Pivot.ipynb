```python
3 #ON VEUT Savoir le nombre de licences de produits que la societe Google a achetees
# pour cela on va redenir notre dataframe  en utilisant seulement les colonnes intervenant dans notre recherche a savoir: Company, Product, Licences
```

```python
licenses = df[['Company', 'Product', 'Licenses']]
licenses
```

```python
# Application de la methode pivot
pd.pivot(data=licenses, index='Company', columns='Product', values='Licenses')
```

```python
# la methode pivot_table
pd.pivot_table(df, index='Company', aggfunc='sum')
```

```python
# en  utilisant la methode groube_by()
df.groupby('Company').sum()
```

```python
df.groupby(['Account Manager', 'Contact'])[['Licenses','Sale Price']].sum()
```

```python
#on peut utiliser la methode pd.pivot_table()
pd.pivot_table(df, index=['Account Manager','Contact'], values='Sale Price', columns='Product', aggfunc=[np.sum, np.max], fill_value=0, margins=True)
```
