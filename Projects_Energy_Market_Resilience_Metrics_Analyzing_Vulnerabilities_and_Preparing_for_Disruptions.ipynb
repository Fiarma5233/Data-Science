```python
# grouping all datasets as dataset


# Grouper tous les ensembles de données sous la variable 'datasets'
datasets = {
    "historical_energy_data": historical_energy_data,  # Ensemble de données sur l'énergie historique
    "infrastructure_data": infrastructure_data,        # Ensemble de données sur l'infrastructure
    "market_data": market_data,                        # Ensemble de données sur le marché
    "regulatory_data": regulatory_data                 # Ensemble de données sur la réglementation
}

# extracting  unique values  and their counts from categorical columns

# Extraction des valeurs uniques et de leurs comptes à partir des colonnes catégorielles
categorical_columns = {
    "historical_energy_data": ['Location/Region', 'Energy Source'],  # Colonnes catégorielles dans 'historical_energy_data'
    "infrastructure_data": ['Infrastructure Status', 'Maintenance Activities', 'Technology Limitations'],  # Colonnes catégorielles dans 'infrastructure_data'
    "market_data": ['Competitor Data', 'Market Trends'],  # Colonnes catégorielles dans 'market_data'
    "regulatory_data": ['Regulatory Changes', 'Compliance Status']  # Colonnes catégorielles dans 'regulatory_data'
}

# Dictionnaire pour stocker les comptes de valeurs catégorielles
categorical_Values_counts = {}

# Boucler sur chaque ensemble de données et ses colonnes catégorielles
for dataset, columns in categorical_columns.items():
    for column in columns:
        # Compter les valeurs uniques dans chaque colonne et les stocker dans le dictionnaire
        categorical_Values_counts[(dataset, column)] = datasets[dataset][column].value_counts()
```

```python
# visualizing
# calculate the number  of rows and columns  for the subplots

num_rows = (len(categorical_Values_counts) + 1) // 2
num_cols = 2

# create a grid of the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 12))

# flatten the axes array for easier  indexing
axes = axes.flatten()

# iterate through  categorical  columns  and plot  them in subplots
# Boucle sur chaque élément dans le dictionnaire 'categorical_Values_counts' avec un index
for i, ((dataset, column), counts) in enumerate(categorical_Values_counts.items()):
    # Sélectionne l'axe de la figure correspondant à l'index actuel
    ax = axes[i]
    # Crée un graphique en barres avec les valeurs des comptes (x) et les catégories (y)
    sns.barplot(x=counts.values, y=counts.index, ax=ax)
    # Définit le titre du graphique
    ax.set_title(f"Value Counts for {dataset} - {column}")
    # Définit l'étiquette de l'axe des x
    ax.set_xlabel('Count')
    # Définit l'étiquette de l'axe des y
    ax.set_ylabel('Column')


# hide any remaining emphy subplots
# Pour chaque index 'i' dans la plage allant du nombre d'éléments dans 'categorical_Values_counts' jusqu'à 'num_rows * num_cols'
for i in range(len(categorical_Values_counts), num_rows * num_cols):
    # Supprime l'axe de la figure correspondant à l'index 'i'
    fig.delaxes(axes[i])



# adjust the layout and display the plots
plt.tight_layout()
plt.show()
plt.close("all")
```

**Energy Demand, Production and Consumption Over Time**

-  We'll start by examining  the trends of energy demand, production and consumption from  the historical_energy_data dataset

- This will give us insights into fluctuations and potential vulnerabilities  in the energy market

```python
# Extract the month and year from 'Date/Time' column
historical_energy_data['Year'] = historical_energy_data['Date/Time'].dt.year
historical_energy_data['Month'] = historical_energy_data['Date/Time'].dt.month


# Create a new column 'Year-Month' for easy plotting
historical_energy_data['Year-Month'] = historical_energy_data['Date/Time'].dt.to_period('M')

# Aggregating data  on a  monthly basis
#monthly_data = historical_energy_data.groupby('Year-Month')[historical_energy_data.select_dtypes(exclude ='object')].mean()
#monthly_data = historical_energy_data.groupby('Year-Month').mean() # Calculate the mean for all numeric columns within each group

numeric_cols = historical_energy_data.select_dtypes(include='number').columns # I add to resolve an aggregation issue
monthly_data = historical_energy_data.groupby('Year-Month')[numeric_cols].mean()

# Setting  figsize
plt.figure(figsize=(15, 6))

# Plotting
sns.lineplot(
    data=monthly_data,
    x=monthly_data.index.astype(str),
    y= 'Energy Production (kWh)',
    label= "Energy Production",
    color = 'gold',
    linestyle = '-',
    linewidth = 1.5,
    ci = None
)

sns.lineplot(
    data=monthly_data,
    x=monthly_data.index.astype(str),
    y= 'Energy Consumption (kWh)',
    label= "Energy Consumption",
    color = 'red',
    linestyle = '-',
    linewidth = 1.5,
    ci = None

)

sns.lineplot(
    data=monthly_data,
    x=monthly_data.index.astype(str),
    y= "Energy Demand" ,
    label= "Energy Demand",
    color = 'green',
    linestyle = '-',
    linewidth = 1.5,
    ci = None

)

# Setting the title and labels
plt.title('Monthly Aggregate of Energy Production , Consumption and Demand Over Time')
plt.xlabel('Date (Year-Month)')
plt.ylabel('Energy (KWh)')
plt.legend(loc= 'upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle = '--', linewidth = 0.5, which= 'both')

# Setting our labels
labels = monthly_data.index.astype(str).tolist()
n = 6
plt.xticks(labels[::n], rotation=360)

plt.tight_layout()
plt.subplots_adjust(hspace=0.5)

# Displaying the plot
plt.show();
plt.close("all")
```

**Market Price and Energy Price Over Time**

-  To understand the pricing dynamic and its relation to market trends

-  Visualization of the Market Price  from the market_data dataset alongside the Energy Price from the historical_energy_data dataset over time

```python
import warnings
warnings.filterwarnings('ignore')

# Merging historical_energy_data and market_data on Date/Time for combined analysis
# Fusionner les données historiques d'énergie avec les données de marché en utilisant la colonne 'Date/Time' comme clé
merged_data = pd.merge(historical_energy_data, market_data, on='Date/Time', how='inner')

# Setting 'Date/Time' as the index  of DataFrame for aggregation
merged_data['Year'] = merged_data['Date/Time'].dt.year
merged_data['Month'] = merged_data['Date/Time'].dt.month

# Aggregating data on a monthly basis
numeric_cols = merged_data.select_dtypes(include='number').columns # I add to resolve an aggregation issue

monthly_merged_data = merged_data.groupby(['Year', 'Month'])[numeric_cols].mean()

# Create a new "Year-Month" column for plotting
# Créer une nouvelle colonne 'Year-Month' en combinant les valeurs des niveaux de l'index

monthly_merged_data['Year-Month'] = monthly_merged_data.index.get_level_values(0).astype(str) + '-' + monthly_merged_data.index.get_level_values(1).astype(str)

# Setting  figsize
plt.figure(figsize=(15, 6))

# Plotting
sns.lineplot(
    data=monthly_merged_data,
    x= "Year-Month",
    y= 'Market Price',
    label= "Market Price",
    color = 'gold',
    linestyle = '-',
    linewidth = 1.5,
    ci = None
)

sns.lineplot(
    data=monthly_merged_data,
    x= "Year-Month",
    y= "Energy Price",
    label= "Energy Price",
    color = 'pink',
    linestyle = '-',
    linewidth = 1.5,
    ci = None

)

# Setting the title and labels
plt.title('Monthly Aggregate of Market Price  and Energy Price  Over Time')
plt.xlabel('Date (Year-Month)')
plt.ylabel('Price ($)')
plt.legend(loc= 'upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle = '--', linewidth = 0.5, which= 'both')

# Setting our labels
labels = monthly_data.index.astype(str).tolist()
n = 6
plt.xticks(labels[::n], rotation=360)

plt.tight_layout()
plt.subplots_adjust(hspace=0.5)

# Displaying the plot
plt.show();
plt.close("all")
```

Correlation between  poor infrastructure status and high technology limitataions and the energy production

```python
# Ajouter une nouvelle colonne 'Poor_Infrastructure' où 1 indique une infrastructure en mauvais état et 0 sinon
infrastructure_data['Poor_Infrastructure'] = infrastructure_data['Infrastructure Status'].apply(lambda x: 1 if x == 'Poor' else 0)

# Ajouter une nouvelle colonne 'High_Technology' où 1 indique une haute technologie et 0 sinon
infrastructure_data['High_Technology'] = infrastructure_data['Technology Limitations'].apply(lambda x: 1 if x == 'High' else 0)

# Créer une nouvelle colonne 'Combined' qui est le produit de 'Poor_Infrastructure' et 'High_Technology'
# Cette colonne sera 1 seulement si les deux conditions (infrastructure en mauvais état et haute technologie) sont vraies
infrastructure_data['Combined'] = infrastructure_data['Poor_Infrastructure'] * infrastructure_data['High_Technology']

# Fusionner les données historiques d'énergie avec les données d'infrastructure sur la colonne 'Date/Time'
merged_data = pd.merge(historical_energy_data, infrastructure_data, on='Date/Time', how='inner')

# Calculer le coefficient de corrélation entre la production d'énergie et la colonne 'Combined'
correlation_coefficient = merged_data['Energy Production (kWh)'].corr(merged_data['Combined'])

# Afficher le coefficient de corrélation
print(f"Correlation coefficient between Energy Production and Combined Infrastructure Status and Technology Limitations: {correlation_coefficient}")
```

**Impact of Regulatory Changes and Operational Costs on Revenue**

-  We will visualize  the relationship  between Regulatory Changes  and the associated Compliance Costs , as well as Operational Costs to the Revenue generated from the company

- Revenue = Energy Consumption * Energy Price

- THis will provide insights into the finantial implications of Regulatory Changes Costs and Operational Cost for Energix Entreprise

```python
# Calculer le revenu en multipliant la consommation d'énergie par le prix de l'énergie
historical_energy_data['Revenue'] = historical_energy_data['Energy Consumption (kWh)'] * historical_energy_data['Energy Price']

# Fusionner les données historiques d'énergie avec les données réglementaires sur la colonne 'Date/Time'
merged_regulatory_data = pd.merge(historical_energy_data, regulatory_data, on='Date/Time', how='inner')

# Extraire l'année à partir de la colonne 'Date/Time' et la stocker dans une nouvelle colonne 'Year'
merged_regulatory_data['Year'] = merged_regulatory_data['Date/Time'].dt.year

# Extraire le mois à partir de la colonne 'Date/Time' et le stocker dans une nouvelle colonne 'Month'
merged_regulatory_data['Month'] = merged_regulatory_data['Date/Time'].dt.month

# Sélectionner uniquement les colonnes numériques pour résoudre un problème d'agrégation
numeric_cols = merged_regulatory_data.select_dtypes(include='number').columns

# Agréger les données sur une base mensuelle en calculant la moyenne des colonnes numériques
monthly_aggregate_data = merged_regulatory_data.groupby('Year-Month')[numeric_cols].mean().reset_index()

# Créer une colonne 'Year-Month' pour faciliter la visualisation, en combinant les colonnes 'Year' et 'Month'
monthly_aggregate_data['Year-Month'] = monthly_aggregate_data['Year'].astype(str) + '-' + monthly_aggregate_data['Month'].astype(str)

plt.figure(figsize=(18, 9))  # Largeur de 18 pouces et hauteur de 9 pouces

# Tracer les coûts opérationnels avec une zone remplie en bleu
plt.fill_between(monthly_aggregate_data['Year-Month'], monthly_aggregate_data['Operational Costs'], color='blue', alpha=0.5, label='Operational Costs')

# Tracer les coûts de conformité en empilant cette zone au-dessus des coûts opérationnels
plt.fill_between(monthly_aggregate_data['Year-Month'],
                 monthly_aggregate_data['Operational Costs'],
                 monthly_aggregate_data['Operational Costs'] + monthly_aggregate_data['Compliance Costs'],
                 color='red', alpha=0.5, label='Compliance Costs')

# Tracer les revenus en tant que ligne verte
sns.lineplot(x='Year-Month', y='Revenue', data=monthly_aggregate_data, color='green', label='Revenue', linewidth=2)

# Ajouter un titre au graphique
plt.title('Impact of Regulatory Changes on Operational, Revenue, and Compliance Costs (Monthly aggregated)')

# Ajouter une étiquette à l'axe des x
plt.xlabel('Year-Month')

# Ajouter une étiquette à l'axe des y
plt.ylabel('Revenue ($)')

# Ajouter une légende en haut à gauche
plt.legend(loc='upper left')

# Ajouter une grille au graphique
plt.grid(True, linestyle='--', linewidth=0.5, which='both')

# Ajuster les étiquettes de l'axe des x pour les afficher tous les 4 mois
n = 4
labels = monthly_aggregate_data['Year-Month'].astype(str).tolist()
plt.xticks(labels[::n], rotation=30)

# Ajuster la disposition du graphique pour éviter les chevauchements
plt.tight_layout()

# Afficher le graphique
plt.show()
```
