```python
# create a funcion that does what me  have done  so far

def prepare_data_and_create_pivot(df1, df2, min_vote = 10000):
  df1, df2 = df1.copy(), df2.copy()
  # merge
  merged = pd.merge(df1, df2, on='movieId', how='inner')
  # Compter les videos par titre
  counter = df['title'].value_counts()
  # selectioner les videos impopulaires ie celles dont le nombre de vote ou likes est < 10000
  rare_movies = values_pd[counter < min_vote].index
  # selectioner les videos populaires ie celles dont le nombre de vote ou likes est >= 10000
  merged_non_rare = merged[~merged.isan(rare_movies)]
  # pivot
  pivottable = merged_non_rare.pivot_table(index=['userId'], columns=['title'], values='rating')
  return pivottable

```

**USER BASED RECOMMENDTIONS**

```python
# select random user

# look at user preference

# select others users with similar viewing  history (60)

#  rank them

# get their other  ratings and figure  out which  movie to recommand
```

```python
# choisir un utilisateur au hasard
random_user = user_movies_df.sample(1, random_state=37).index[0]
```

```python
random_user
```

```python

#DataFrame random_user_df qui contient uniquement les films regardés par l'utilisateur identifié par random_user.

random_user_df = user_movies_df[user_movies_df.index == random_user]
```

```python
random_user_df
```

```python
 # suppression des videos non regardees ie NaN
 movies_watched = random_user_df.dropna(axis=1).columns.tolist()
```

```python
movies_watched
```

```python
#DataFrame movies_watched_df qui contient uniquement les films que l'utilisateur a regardés, en se basant sur la liste de conditions movies_watched
movies_watched_df = user_movies_df[movies_watched]
```

```python
'''
ce code compte le nombre de films regardés par chaque utilisateur en transposant le DataFrame,
en vérifiant les valeurs non nulles, puis en additionnant ces valeurs pour chaque utilisateur.
'''
user_movies_count = movies_watched_df.T.notnull().sum()
```

```python
user_movies_count
```

```python
user_movies_count.max()
```

```python
len(movies_watched)
```

```python
# Filtre les utilisateurs qui ont regardé plus de 60% des films
# movies_watched_df.shape[1] retourne le nombre total de films (nombre de colonnes)
# On calcule 60% du nombre total de films
# user_movies_count > (movies_watched_df.shape[1] * 60) / 100 crée une condition booléenne pour chaque utilisateur
# qui vérifie s'ils ont regardé plus de 60% des films
# On utilise cette condition pour filtrer user_movies_count et garder uniquement les utilisateurs qui remplissent cette condition
# .index retourne les index des utilisateurs filtrés
user_same_movies = user_movies_count[user_movies_count > (movies_watched_df.shape[1] * 60) / 100].index
```

```python
user_same_movies
```

```python
# Filtrer le DataFrame pour ne garder que les lignes (films) regardées par les utilisateurs qui ont regardé plus de 60% des films
# movies_watched_df.index.isin(user_same_movies) crée une série de booléens indiquant si chaque index (utilisateur)
# est dans la liste user_same_movies (les utilisateurs qui ont regardé plus de 60% des films)
# movies_watched_df[movies_watched_df.index.isin(user_same_movies)] utilise cette série de booléens pour filtrer le DataFrame
# et ne garder que les lignes dont l'index est dans user_same_movies
filtered_df = movies_watched_df[movies_watched_df.index.isin(user_same_movies)]
```

```python
filtered_df
```

```python
# Calcule la corrélation entre les utilisateurs pour les films regardés et la convertit en série
corr_df = filtered_df.T.corr().unstack()
```

```python
# Récupère les corrélations entre l'utilisateur spécifié (random_user) et tous les autres utilisateurs
corr_df[random_user]
```

```python
# Crée un DataFrame avec les utilisateurs ayant une corrélation supérieure à 0.65 avec l'utilisateur spécifié (random_user),
# en utilisant les valeurs de corrélation comme colonne 'corr'
top_users = pd.DataFrame(corr_df[random_user][corr_df[random_user] > 0.65], columns=['corr'])
```

```python
top_users
```

```python
# Fusionne les utilisateurs ayant une haute corrélation avec random_user avec les évaluations des films pour ces utilisateurs
# en utilisant 'userId' comme clé de fusion
top_users_rating = pd.merge(top_users, rating[['userId', 'movieId', 'rating']], how='inner', on='userId')
```

```python
top_users_rating
```

```python
# Calcule le rating pondéré pour chaque utilisateur en multipliant la corrélation par la note du film
top_users_rating['weighted_rating'] = top_users_rating['corr'] * top_users_rating['rating']
```

```python
top_users_rating
```

```python
# Crée un tableau croisé dynamique pour calculer la moyenne des ratings pondérés par film
recommendation_df = top_users_rating.pivot_table(index=['movieId'], values='weighted_rating', aggfunc='mean')
```

```python
recommendation_df
```

```python
# Filtre les films avec un rating pondéré supérieur à 3.5, les trie par rating pondéré décroissant,
# et sélectionne les 10 meilleurs films à recommander
movies_to_be_recommended = recommendation_df[recommendation_df['weighted_rating'] > 3.5].sort_values(by='weighted_rating', ascending=False).head(10)
```

```python
movies_to_be_recommended
```

```python
# Récupère les titres des films dont les identifiants figurent dans la liste des films recommandés
movie['title'][movie['movieId'].isin(movies_to_be_recommended.index)]
```

**Item-Based Recommendation**

```python
# random user
user = user_movies_df.sample(1, random_state=37).index[0]
user

# get his  last seen movie , that was rated highly

# see what  everyone  else  felt about  the movie

# get  another movie  with  the same  effect
```

```python
# Sélectionne l'identifiant du deuxième film le mieux noté (rating = 5) par un utilisateur spécifique, trié par date d'évaluation
pick = rating[(rating['rating'] == 5) & (rating['userId'] == user)].sort_values(by='timestamp', ascending=False).iloc[1]['movieId']
```

```python
# Récupère le titre du film correspondant à l'identifiant sélectionné
picked_movie_name = movie[movie['movieId'] == pick]['title'].iloc[0]
```

```python
picked_movie_name
```

```python
filtered_movies = user_movies_df[picked_movie_name]
```

```python
filtered_movies
```

```python
# Supprime les données de l'utilisateur spécifié et les films déjà regardés du DataFrame
users_wo_user = user_movies_df.drop(user, axis=0).drop(movies_watched, axis=1)
```

```python
users_wo_user
```

```python
# Supprime les données de l'utilisateur spécifié du DataFrame des films filtrés
filtered_wo_user = filtered_movies.drop(user, axis=0)
```

```python
filtered_wo_user
```

```python
# Calcule la similarité entre les films restants en utilisant la corrélation avec les films filtrés, trie les résultats par similarité décroissante
movies_similarity = users_wo_user.corrwith(filtered_wo_user).sort_values(ascending=False).reset_index()
```

```python
movies_similarity.columns = ['title', 'similarity']
```

```python
movies_similarity.sort_values(by='similarity', ascending=False).iloc[1:6]
```

**Hybrid Recommendation**

```python
movies_similarity
```

```python
recommendation_df
```

```python
# Fusionne le DataFrame de recommandations avec les informations sur les films, puis sélectionne les colonnes pertinentes pour obtenir un tableau avec les films, leurs ratings pondérés et leurs titres
movies_ordered_by_rating = pd.merge(recommendation_df, movie, how='inner', on='movieId')[['movieId', 'weighted_rating', 'title']]
```

```python
movies_ordered_by_rating
```

```python
# Fusionne les données de similarité des films avec les informations sur les films ordonnés par rating, en utilisant les titres de films comme clé de fusion
merged = pd.merge(movies_similarity, movies_ordered_by_rating, how='inner', on='title')
```

```python
merged
```

```python
merged['hybrid'] = merged['similarity'] * merged['weighted_rating']
```
