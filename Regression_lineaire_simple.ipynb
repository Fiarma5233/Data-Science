```python
#Prediction de la vente en fonction de la depense publicitaire
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/MasterClass_DataScience_Machine_Learning_Notebooks_et_Datasets/08-Modèles-de-Régression-Linéaire/Datasets/Advertising.csv')
```

```python
df
```

```python
'''
 Construction de la  droite de regression lineaire simple
 c'est a dire avec une seule colonne ou caracteristique
 Pour cela, on va creer une nouvelle colonne 'depense_puplicitaire' qui sera  la somme des depenss a la tele, radio et journaux



 '''
```

```python
df['depense_pub'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df
```

```python
# nuage de points ou scatterplot()
# construction de la droite de regression lineaire simple: y = ax + b  en se basant sur la methode des moindres carrees
# avec la fonction regplot() de sns

fig = plt.figure(figsize=(10,5), dpi=200)
sns.regplot(df, x='depense_pub', y='sales')
```

```python
X = df['depense_pub'] # correspond a notre caracteristique ou colonne
y = df['sales'] # correspond a notre label ou variable cible ( celle a predire)
```

```python
''' la fonction polyfit() nous permet de calculer la pente et la constante de notre droite de regression lineaire y = ax + b
                            avec a = pente
                                b = la constante
                                y = notre prediction


 '''
```

```python
help(np.polyfit)
```

```python
np.polyfit(X, y, deg=1)
```

```python
potential_spend = np.linspace(0,500,100) # depense potentielles comprises entre 0 et 500 espacees de 100

predicted_sales = 0.04868788 * potential_spend + 4.24302822
```

```python

 # on obtient la meme droite de regression lineaire simple en  faisant ceci
fig = plt.figure(figsize=(10,5), dpi=200)
sns.scatterplot(df, x='depense_pub', y='sales')
plt.plot(potential_spend, predicted_sales, color='red')
```
