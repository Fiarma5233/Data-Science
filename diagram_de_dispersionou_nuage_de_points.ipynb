```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/MasterClass_DataScience_Machine_Learning_Notebooks_et_Datasets/05-Seaborn/Datasets/dm_office_sales.csv')
df
```

```python
from posixpath import abspath
# quelle est la relation entre le salaire et la vente
# Pour resoudre cela  nous allons visualiser le nuage de ppoint

fig = plt.figure(figsize=(12,4), dpi=200)

sns.scatterplot(x='salary', # abscisses correspondant au salaire
                y='sales', # ordonnees qui correspont aux ventes
                data=df,  # jeux de donnees
                hue='level of education',  # Pour regler les teintes
                palette='spring',  # pour une palette de couleur : https://matplotlib.org/stable/users/explain/colors/colormaps.html
                s=100, # pour agrandir tous les points sur le diagramme
                alpha=0.2, # pour la transpararence des points ; valeur comprise entre 0 et 1
                style='level of education',# qui correspond au marker et doit etre une colone categorielle
                );

plt.savefig("Nuage de point avec scatterplot.png")
```
